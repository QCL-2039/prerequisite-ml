```python
#IM Task-3

arr=rand.integers(1,10,10)
print("First array:\n",arr)

arr2=rand.integers(10,20,10)
print("Second array:\n",arr2)

arr3=rand.integers(30,40,10)
print(arr3)
print()

final_arr=np.vstack((arr,arr2,arr3))
print(final_arr)

final_arr=np.hstack((arr,arr2,arr3))
print(final_arr)
```

```python
arr = rand.integers(-10, 10, size=(3, 3))
print("Original:\n", arr)

arr[arr < 0] = 0
print("Modified:\n", arr)

arr = rand.integers(-10, 10, size=(3,3))
print("Original:\n", arr)

arr = np.where(arr < 0, 0, arr)
print("Modified:\n", arr)
```

```python
mat=np.matrix(arr)
dia_mat=np.diag(mat)
print(dia_mat)
```
