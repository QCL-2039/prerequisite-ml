```python
import pandas as pd
import numpy as np
```

| Operator | Meaning        | Method Equivalent |
| -------- | -------------- | ----------------- |
| `+`      | Addition       | `.add()`          |
| `-`      | Subtraction    | `.sub()`          |
| `*`      | Multiplication | `.mul()`          |
| `/`      | Division       | `.div()`          |
| `//`     | Floor division | `.floordiv()`     |
| `%`      | Modulus        | `.mod()`          |
| `**`     | Power          | `.pow()`          |


```python
x=np.random.randint(50,100,10)
y=np.random.randint(50,100,10)

df=pd.DataFrame({"X":x,
                 "Y":y})

df["Addition of X+Y"]=df["X"]+df["Y"]
df["Subtraction of X-Y"]=df["X"]-df["Y"]
df["Multiplication of X*Y"]=df["X"]*df["Y"]
df["Divition of X/Y"]=df["X"]/df["Y"]

df
```

```python
df["X>y"]=df["X"]>df["Y"]
df
df["X & y"]=df["X"] & df["Y"]
df
```
