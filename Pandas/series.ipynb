```python
import pandas as pd
import numpy as np

print("Version of your numpy is:",np.__version__)
print("Version of your pandas is:",pd.__version__)
```

| Property    | Why                 |
| ----------- | ------------------- |
| `dtype`     | Model compatibility |
| `values`    | NumPy & ML          |
| `index`     | Data alignment      |
| `hasnans`   | Missing values      |
| `shape`     | Data size           |
| `nunique()` | Feature analysis    |
nunique() telll the number of unique data


```python
rand=np.random.default_rng(seed=42)

x=rand.integers(40,100,10)

y=np.arange(1,11)

data=pd.Series(x,y,dtype="f")

print("Values in the series:",data.values)
print("Type of the values:",type(data.values))

print("Size of the data:",data.size)
print("Shape of the data:",data.shape)
print("Dimensions of the data:",data.ndim)
print("Has there any nan values:",data.hasnans)
print("Check is the values are unique?:",data.is_unique)
print("Check out the memory usage by this data set:",data.memory_usage())
print("Memory usage by data is:",data.memory_usage(index=False))
print("Number of unique data are:",data.nunique())
print("Sum of data:",np.sum(data))
print("Variance of the data:",np.var(data))
print("Standard Deviatiaon:",np.std(data))
print("Mean of the data:",np.mean(data))
print("Median of data",data.median())
print("Mode of the data:",data.mode())
```

```python
x=rand.integers(40,100,10)

y=np.arange(1,11)

z=np.arange(1,11)

data1=pd.Series(x,y)
data2=pd.Series(x,z)

print(data1+data2)#if the index are same then the value will be added,otherwise nan
```
