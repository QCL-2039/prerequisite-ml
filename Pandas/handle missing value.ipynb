| Parameter | Purpose                 |
| --------- | ----------------------- |
| `axis`    | Rows (0) or Columns (1) |
| `how`     | any / all               |
| `subset`  | Specific columns        |
| `thresh`  | Min non-null values     |
| `inplace` | Modify directly         |


```python
df.dropna()
df.dropna(axis=1)
df.dropna(subset=["Col1","Col2"])
df.dropna(how="any")
df.dropna(thresh=1)
```

| Parameter   | Meaning                           |
| ----------- | --------------------------------- |
| `axis=1`    | Work on columns                   |
| `subset`    | Check specific columns            |
| `how="any"` | Drop if at least one NaN          |
| `thresh=n`  | Keep rows with ≥ n non-NaN values |


```python
df.dropna(
    axis=0,
    how='any',
    thresh=None,
    subset=None,
    inplace=False
)
```

```python
df.dropna()
# Drops all ROWS that contain at least one missing (NaN) value.
# axis=0 and how='any' are used by default.

df.dropna(axis=1)
# Drops all COLUMNS that contain at least one missing (NaN) value.

df.dropna(subset=["Col1", "Col2"])
# Drops rows ONLY if Col1 or Col2 has missing values.
# Other columns are ignored when checking for NaN.

df.dropna(how="any")
# Drops rows if ANY column in the row has a missing value.
# This is the default behavior of dropna().

df.dropna(thresh=1)
# Keeps rows that have at least 1 non-missing (non-NaN) value.
# Rows with all values missing are dropped.
```
