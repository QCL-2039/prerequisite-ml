```python
import numpy as np
import pandas as pd
```

```python

students = pd.DataFrame({
    "student_id": [1,2,3,4,5],
    "name": ["Rahim","Karim","Salma","Nila","Hasan"],
    "department": ["CSE","EEE","CSE","BBA","CSE"]
})

scores = pd.DataFrame({
    "student_id": [1,2,3,4,5],
    "subject": ["Math","Physics","Math","Math","Math"],
    "score": [85,78,88,92,75]
})
students.join(scores)
students.join(
    scores,
    how="outer",
    lsuffix="_left",
    rsuffix="_right"
)
```
