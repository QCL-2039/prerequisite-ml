| Property     | Purpose                  |
| ------------ | ------------------------ |
| `df.shape`   | Rows & columns count     |
| `df.size`    | Total elements           |
| `df.ndim`    | Dimension (always 2)     |
| `df.columns` | Column labels            |
| `df.index`   | Row labels               |
| `df.dtypes`  | Data type of each column |
| `df.values`  | Data as NumPy array      |
| `df.empty`   | Check if empty           |
| `df.T`       | Transpose                |


| Method          | Use                   |
| --------------- | --------------------- |
| `df.head()`     | First 5 rows          |
| `df.tail()`     | Last 5 rows           |
| `df.info()`     | Memory, nulls, dtypes |
| `df.describe()` | Statistical summary   |
| `df.sample()`   | Random rows           |
