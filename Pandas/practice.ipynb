```python
import pandas as pd
import numpy as np

data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Math": [85, 90, np.nan, 70, 88],
    "Physics": [np.nan, 92, 75, 80, np.nan],
    "Chemistry": [78, np.nan, 82, 76, 85],
    "Attendance": [90, 85, np.nan, 95, 88]
}

df = pd.DataFrame(data)
print(df)
```

```python
df[df["Math"]>80]
print(df)
```

```python
df.isnull()
df.nunique()
df.fillna(0)
```
