```python
import matplotlib.pyplot as plt   # Import Matplotlib for plotting

x = [1, 2, 3, 4]                  # X-axis positions of bars
y = [5, 6, 7, 8]                  # Heights of the bars
c = ["r", "g", "b", "y"]          # Colors for each bar (r=red, g=green, b=blue, y=yellow)

plt.bar(x, y, color=c)            # Create a bar chart with custom colors
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Bar Chart with Custom Color")
plt.show()                        # Display the plot window
```

```python
language=["Python","Java","C++","C"]
per_24=[80,70,60,50]
per_25=[90,60,70,40]

plt.title("Programming Language Preferences",fontsize=10)
plt.xlabel("X-axis",fontsize=10)
plt.ylabel("Y-axis",fontsize=10)
plt.bar(language,per_24,label="Per24")
plt.bar(language,per_25,label="Per25")
plt.legend()
plt.show()
```

```python
language=["Python","Java","C++","C"]
per_24=[80,70,60,50]
per_25=[90,60,70,40]

p=np.arange(len(language))
width=.4
p1=[j+width for j in p]
plt.figure(figsize=(12,6))
plt.title("Programming Language Preferences",fontsize=10)
plt.xlabel("X-axis",fontsize=10)
plt.ylabel("Y-axis",fontsize=10)
plt.bar(p,per_24,label="Per24",width=.4)
plt.bar(p1,per_25,label="Per25",width=.4)
plt.legend()
plt.tight_layout()

plt.xticks(p+width/2,language)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

languages = [
    "Python",
    "JavaScript",
    "Java",
    "C#",
    "C++",
    "Go",
    "Rust",
    "TypeScript",
    "Kotlin",
    "Swift"
]

width = 0.3
p = np.arange(len(languages))
p1 = p + width

preferences_2024 = [30, 18, 15, 10, 8, 5, 4, 5, 3, 2]
preferences_2025 = [28, 17, 14, 11, 7, 7, 6, 5, 3, 2]

plt.figure(figsize=(12, 6))
plt.bar(p, preferences_2024, width=width, label="2024")
plt.bar(p1, preferences_2025, width=width, label="2025")

plt.legend()
plt.xticks(p + width/2, languages, rotation=45)
plt.title("Programming Language Preferences (2024 vs 2025)", fontsize=14)
plt.xlabel("Programming Languages", fontsize=12)
plt.ylabel("Preference %", fontsize=12)

plt.tight_layout()
plt.show()
```
