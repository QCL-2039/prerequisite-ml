```python
import matplotlib.pyplot as plt

import numpy as np

rand=np.random.default_rng()
```

```python
languages = [
    "Python",
    "JavaScript",
    "Java",
    "C#",
    "C++",
    "Go",
    "Rust",
    "TypeScript",
    "Kotlin",
    "Swift"
]
preferences_2024 = [30, 18, 15, 10, 8, 5, 4, 5, 3, 2]
preferences_2025 = [28, 17, 14, 11, 7, 7, 6, 5, 3, 2]

x=[i for i in range(len(preferences_2025))]


plt.figure(figsize=(12,6))
plt.suptitle("Sub Plotting Example")

plt.subplot(2,2,1)
plt.title("Linear Plotting")
plt.plot(x,preferences_2025)
plt.xticks(x,languages,rotation=45)

plt.subplot(2,2,2)
plt.title("Bar Graph")
plt.bar(x,preferences_2025)
plt.xticks(x,languages,rotation=45)

plt.subplot(2,2,3)
plt.title("Scatter Plotting")
plt.scatter(x,preferences_2025)
plt.xticks(x,languages,rotation=45)

plt.subplot(2,2,4)
plt.title("Pie Chart")
plt.pie(preferences_2025,labels=languages,autopct="%0.1f%%",radius=1.5)

plt.savefig("Sub Plotting",dpi=2000)
plt.tight_layout()
plt.show()
```

```python
languages = [
    "Python",
    "JavaScript",
    "Java",
    "C#",
    "C++",
    "Go",
    "Rust",
    "TypeScript",
    "Kotlin",
    "Swift"
]
preferences_2024 = [30, 18, 15, 10, 8, 5, 4, 5, 3, 2]
preferences_2025 = [28, 17, 14, 11, 7, 7, 6, 5, 3, 2]

x=[i for i in range(len(preferences_2025))]


plt.figure(figsize=(12,6))
plt.suptitle("Sub Plotting Example")

plt.subplot(2,2,1)
plt.title("Linear Plotting")
plt.plot(x,preferences_2025)
plt.xticks(x,languages,rotation=45)

plt.subplot(2,2,2)
plt.title("Bar Graph")
plt.bar(x,preferences_2025)
plt.xticks(x,languages,rotation=45)

plt.subplot(2,2,3)
plt.title("Scatter Plotting")
plt.scatter(x,preferences_2025)
plt.xticks(x,languages,rotation=45)

plt.subplot(2,2,4)
plt.title("Pie Chart")
plt.pie(preferences_2025,labels=languages,autopct="%0.1f%%",radius=1.5,wedgeprops={"edgecolor":"r"},shadow=True)

# plt.savefig("Sub Plotting-1",dpi=2000,facecolor="r")
plt.tight_layout()
plt.show()
```
