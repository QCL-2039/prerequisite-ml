```python
import matplotlib.pyplot as plt
import numpy as np
rand=np.random.default_rng()
import random
```

```python

num=rand.integers(low=1,high=100,size=10)
num2=num

plt.figure(figsize=(12,6))
plt.title("Pie Chart")
plt.pie(num,labels=num2)
plt.tight_layout()
plt.show()
```
