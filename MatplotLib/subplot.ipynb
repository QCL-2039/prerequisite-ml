```python
import matplotlib.pyplot as plt
import numpy as np

languages = [
    "Python", "JavaScript", "Java", "C#", "C++",
    "Go", "Rust", "TypeScript", "Kotlin", "Swift"
]

preferences_2024 = [30, 18, 15, 10, 8, 5, 4, 5, 3, 2]
preferences_2025 = [28, 17, 14, 11, 7, 7, 6, 5, 3, 2]

# Convert languages to numeric positions
x = np.arange(len(languages))

plt.figure(figsize=(12, 8))
plt.suptitle("Sub Plotting", fontsize=16, fontweight="bold")

# 1️⃣ LINE PLOT
plt.subplot(3, 2, 1)
plt.title("Linear Plot")
plt.plot(x, preferences_2025, marker='o')
plt.xticks(x, languages, rotation=45)

# 2️⃣ BAR CHART
plt.subplot(3, 2, 2)
plt.title("Bar Graph")
plt.bar(x, preferences_2025)
plt.xticks(x, languages, rotation=45)

# 3️⃣ PIE CHART
plt.subplot(3, 2, 3)
plt.title("Pie Chart")
plt.pie(preferences_2025, labels=languages, autopct="%1.1f%%")

# 4️⃣ SCATTER PLOT
plt.subplot(3, 2, 4)
plt.title("Scatter Plot")
plt.scatter(x, preferences_2025)
plt.xticks(x, languages, rotation=45)

# 5️⃣ HISTOGRAM
plt.subplot(3, 2, 5)
plt.title("Histogram Plot")
plt.hist(preferences_2025, bins=5)

# 6️⃣ STEM PLOT
plt.subplot(3, 2, 6)
plt.title("Stem Plot")
plt.stem(x, preferences_2025)
plt.xticks(x, languages, rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
