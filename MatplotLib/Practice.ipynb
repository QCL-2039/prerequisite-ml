```python
import matplotlib.pyplot as plt

counts = [5, 3, 9, 2]
labels = ['A', 'B', 'C', 'D']

plt.figure()
bars = plt.bar(labels, counts)


# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X position
        height,                             # Y position
        str(height),                        # Label text
        ha='center', va='bottom'
    )

plt.xlabel("Labels")
plt.ylabel("Counts")
plt.title("Bar Chart with Value Labels")
plt.show()
# print(type(bars))
# print(bars[0])
```

```python
import matplotlib.pyplot as plt

shares = [40, 25, 20, 15]
labels = ['P', 'Q', 'R', 'S']

# Find the index of the largest slice
pos_max = shares.index(max(shares))

# Create explode list: 0 for all except the largest slice
explode = [0.1 if i == pos_max else 0 for i in range(len(shares))]

# Plot pie chart
plt.pie(shares, labels=labels, autopct="%0.1f%%", explode=explode,shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

langs = ['Py', 'JS', 'C++']
scores = [30, 20, 10]

# Sort by scores (descending)
langs_sorted = [x for _, x in sorted(zip(scores, langs), reverse=True)]
scores_sorted = sorted(scores, reverse=True)

# Plot horizontal bar chart
plt.barh(langs_sorted, scores_sorted)
plt.xlabel("Scores")
plt.ylabel("Languages")
plt.title("Language Scores (Sorted Descending)")
plt.show()
```

```python
#Question 11

languages = [
    "Python",
    "JavaScript",
    "Java",
    "C#",
    "C++",
    "Go",
    "Rust",
    "TypeScript",
    "Kotlin",
    "Swift"
]
preferences_2024 = [30, 18, 15, 10, 8, 5, 4, 5, 3, 2]
preferences_2025 = [28, 17, 14, 11, 7, 7, 6, 5, 3, 2]

x=np.arange(len(languages))
x1=x+.3

plt.figure(figsize=(12,6))
plt.title("Bar Chart")
plt.xlabel("Programming Languages")
plt.ylabel("Languages Preferences")
plt.bar(x,preferences_2024,label="%24",width=.3)
plt.bar(x1,preferences_2025,label="%2025",width=.3)
plt.xticks(x+.3/2,languages,rotation=45)

plt.legend()
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
days = [1, 2, 3, 4, 5, 6, 7]
temperature = [30, 32, 31, 29, 28, 27, 26]     # Left y-axis
precipitation = [5, 10, 3, 0, 8, 12, 6]         # Right y-axis

fig, ax1 = plt.subplots()

# Left axis → Temperature
ax1.plot(days, temperature, color='red', marker='o')
ax1.set_xlabel("Day")
ax1.set_ylabel("Temperature (°C)", color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Right axis → Precipitation
ax2 = ax1.twinx()
ax2.plot(days, precipitation, color='blue', marker='s')
ax2.set_ylabel("Precipitation (mm)", color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title("Temperature and Precipitation vs Day")
plt.show()
```
