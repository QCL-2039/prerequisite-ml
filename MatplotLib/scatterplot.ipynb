```python
import matplotlib.pyplot as plt
import numpy as np
import random
rand=np.random.default_rng()
```

```python
even_num=np.arange(2,22,2)
squre_num=even_num**2
print(even_num)
print(squre_num)

plt.figure(figsize=(12,6))
plt.plot(even_num,squre_num)
plt.tight_layout()
plt.show()
```

```python
even_num=np.arange(2,22,2)
squre_num=even_num**2
print(even_num)
print(squre_num)

colors=['r','g','b','r','g','b','r','g','b','r']
sizes=rand.integers(40,50,10)
plt.figure(figsize=(12,6))
plt.title("Polynomial Graph of Square")
plt.xlabel("Integers value",fontsize=15)
plt.ylabel("Square values",fontsize=15)

plt.scatter(even_num,squre_num,c=colors,s=sizes)
plt.tight_layout()
plt.show()
```

```python
even_num=np.arange(2,22,2)
squre_num=even_num**2
print(even_num)
print(squre_num)

colors=['r','g','b','r','g','b','r','g','b','r']
sizes=rand.integers(40,50,10)
plt.figure(figsize=(12,6))
plt.title("Polynomial Graph of Square")
plt.xlabel("Integers value",fontsize=15)
plt.ylabel("Square values",fontsize=15)

plt.scatter(even_num,squre_num,c=sizes,cmap="viridis")
t=plt.colorbar()
t.set_label("Color Bar",fontsize=20)

plt.tight_layout()
plt.show()
```

```python
even_num=np.arange(2,22,2)
squre_num=even_num**2
quad_num=even_num**3
print(even_num)
print(squre_num)

colors=['r','g','b','r','g','b','r','g','b','r']
sizes=rand.integers(40,50,10)
plt.figure(figsize=(12,6))
plt.title("Polynomial Graph of Square")
plt.xlabel("Integers value",fontsize=15)
plt.ylabel("Square values",fontsize=15)

plt.scatter(even_num,squre_num,c=sizes,cmap="viridis")
plt.scatter(even_num,quad_num,c=sizes)
t=plt.colorbar()
t.set_label("Color Bar",fontsize=20)

plt.tight_layout()
plt.show()
```
