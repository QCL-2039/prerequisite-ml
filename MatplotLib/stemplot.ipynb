```python
import matplotlib.pyplot as plt
import numpy as np
import random

rand=np.random.default_rng()
```

```python
languages = [
    "Python",
    "JavaScript",
    "Java",
    "C#",
    "C++",
    "Go",
    "Rust",
    "TypeScript",
    "Kotlin",
    "Swift"
]
preferences_2024 = [30, 18, 15, 10, 8, 5, 4, 5, 3, 2]
preferences_2025 = [28, 17, 14, 11, 7, 7, 6, 5, 3, 2]

plt.figure(figsize=(12,6))
plt.title("Stem Plot")
plt.xlabel("Programming Languages")
plt.ylabel("Languages Preferences")
# plt.bar(languages,preferences_2024)
plt.stem(languages,preferences_2024)
plt.show()
```
