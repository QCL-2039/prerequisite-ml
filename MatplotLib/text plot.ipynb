```python
import numpy as np
import matplotlib.pyplot as plt

languages = [
    "Python",
    "JavaScript",
    "Java",
    "C#",
    "C++",
    "Go",
    "Rust",
    "TypeScript",
    "Kotlin",
    "Swift"
]

preferences_2024 = [30, 18, 15, 10, 8, 5, 4, 5, 3, 2]
preferences_2025 = [28, 17, 14, 11, 7, 7, 6, 5, 3, 2]

x=[i for i in range(len(languages))]

plt.figure(figsize=(12,6))

plt.title("Texting on the graph")

plt.scatter(languages,preferences_2025)
plt.xticks(languages,x)
plt.yticks(preferences_2025)

plt.annotate("Java",xy=(4,7),xytext=(5,10),arrowprops=dict(facecolor="black",shrink=100))
plt.show()
```
