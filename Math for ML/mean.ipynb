#Why ML cares about the mean
#Mean tells the model where the center of the data lies.

#Scaling (standardization) uses the mean:
#Sensitive to outliers

```python
import numpy as np
import matplotlib.pyplot as plt

data_set_without_outliers = np.array([44, 47, 44, 42, 49, 42, 43, 43, 41, 46])
data_set_with_outliers = np.array([45, 46, 49, 43, 40, 49, 48, 47, 46, 100])

print("Mean of data set without outliers:", data_set_without_outliers.mean())
print("Mean of data set with outliers:", data_set_with_outliers.mean())

plt.figure(figsize=(12, 6))

# Plot 1: Without outliers
plt.subplot(1, 2, 1)
plt.title("Mean without Outliers")
plt.scatter(range(len(data_set_without_outliers)), data_set_without_outliers, c="black")
plt.scatter(len(data_set_without_outliers)//2, data_set_without_outliers.mean(),
            c="red", marker="x", s=120, label="Mean")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

# Plot 2: With outliers
plt.subplot(1, 2, 2)
plt.title("Mean with Outliers")
plt.scatter(range(len(data_set_with_outliers)), data_set_with_outliers, c="blue")
plt.scatter(len(data_set_with_outliers)//2, data_set_with_outliers.mean(),
            c="red", marker="x", s=120, label="Mean")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

plt.tight_layout()
plt.show()
```
