```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
data_set_without_outliers = pd.Series([44, 47, 44, 42, 49, 42, 43, 43, 41, 46])
data_set_with_outliers = pd.Series([45, 46, 49, 43, 40, 49, 48, 47, 46, 100])

# Calculate modes
mode_without = data_set_without_outliers.mode()[0]   # first mode if multiple
mode_with = data_set_with_outliers.mode()[0]

print("Mode (without outliers):", mode_without)
print("Mode (with outliers):", mode_with)

plt.figure(figsize=(12, 6))

# Subplot 1: Without outliers
plt.subplot(1, 2, 1)
plt.title("Mode without Outliers")
plt.scatter(range(len(data_set_without_outliers)), data_set_without_outliers, c="black")
plt.axhline(mode_without, color="red", linestyle="--", label=f"Mode = {mode_without}")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

# Subplot 2: With outliers
plt.subplot(1, 2, 2)
plt.title("Mode with Outliers")
plt.scatter(range(len(data_set_with_outliers)), data_set_with_outliers, c="blue")
plt.axhline(mode_with, color="red", linestyle="--", label=f"Mode = {mode_with}")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

plt.tight_layout()
plt.show()
```
