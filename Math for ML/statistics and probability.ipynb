```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random 
import math
```

```python

rand = np.random.default_rng(seed=42)
x = rand.integers(50, 70, 20)

meanX = x.mean()

# Absolute deviations
abs_dev = np.abs(x - meanX)

data=pd.DataFrame({"Raw Data":x,
                   "Absolute Deviation":abs_dev})
print(data)
# Mean Deviation
mean_dev = abs_dev.mean()
print("Mean =", meanX)
print("Mean Deviation =", mean_dev)
print("Variance:",x.var(ddof=1))
print("Standard Deviation:",x.std(ddof=1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Group 1: Low variance (values close together)
group1 = np.array([48, 50, 52, 49, 51])

# Group 2: High variance (values far apart)
group2 = np.array([30, 50, 70, 20, 90])

print("Mean of both groups:")
print("Group1 mean:", group1.mean())
print("Group2 mean:", group2.mean())

print("\nVariance:")
print("Group1 variance:", group1.var())
print("Group2 variance:", group2.var())

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.scatter(range(len(group1)), group1, c="green")
plt.title("Low Variance Group")
plt.ylim(0,100)

plt.subplot(1,2,2)
plt.scatter(range(len(group2)), group2, c="red")
plt.title("High Variance Group")
plt.ylim(0,100)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

group1 = np.array([40, 45, 50, 55, 60])   # mean = 50
group2 = np.array([70, 75, 80, 85, 90])   # mean = 80

print("Group1 Mean:", group1.mean())
print("Group2 Mean:", group2.mean())

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.scatter(range(len(group1)), group1, c="blue")
plt.title("Group 1 (Mean ≈ 50)")

plt.subplot(1,2,2)
plt.scatter(range(len(group2)), group2, c="orange")
plt.title("Group 2 (Mean ≈ 80)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np

group = np.array([10, 12, 14, 15, 16])   # No outlier
group_outlier = np.array([10, 12, 14, 15, 200])  # Outlier added

print("Median without outlier:", np.median(group))
print("Median with outlier:", np.median(group_outlier))
print("Mean without outlier:", group.mean())
print("Mean with outlier:", group_outlier.mean())
```
