```python
import numpy as np
import matplotlib.pyplot as plt

# Use np.sort() to avoid the None issue
data_set_without_outliers = np.sort(np.array([44, 47, 44, 42, 49, 42, 43, 43, 41, 46]))
data_set_with_outliers = np.sort(np.array([45, 46, 49, 43, 40, 49, 48, 47, 46, 100]))

median_without = np.median(data_set_without_outliers)
median_with = np.median(data_set_with_outliers)

print("Median (without outliers):", median_without)
print("Median (with outliers):", median_with)

plt.figure(figsize=(12, 6))

# Plot 1: Without outliers
plt.subplot(1, 2, 1)
plt.title("Median without Outliers")
plt.scatter(range(len(data_set_without_outliers)), data_set_without_outliers, c="black")
plt.axhline(median_without, color="red", linestyle="--", label="Median")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

# Plot 2: With outliers
plt.subplot(1, 2, 2)
plt.title("Median with Outliers")
plt.scatter(range(len(data_set_with_outliers)), data_set_with_outliers, c="blue")
plt.axhline(median_with, color="red", linestyle="--", label="Median")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

plt.tight_layout()
plt.show()
```
