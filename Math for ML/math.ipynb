```python
import numpy as np
import matplotlib.pyplot as plt

rand=np.random.default_rng(seed=42)
groupA = rand.integers(50, 70, size=200)
groupB = rand.integers(50, 70, size=200)

print("Data of group A:\n",groupA)
print()
maxA=max(groupA)
minA=min(groupA)
print("Maximum and minimum groupA element:",maxA,minA)
meanA = groupA.mean()

print("Data of groupB:\n",groupB)
print()
maxB=max(groupB)
minB=min(groupB)
print("Maximum and minimum groupB element:",maxB,minB)
print("Mean value of groupA:",meanA)
meanB = groupB.mean()
print("Mean value of groupB:",meanB)

plt.figure(figsize=(12, 6))

# Plot 1: Group A
plt.subplot(1, 2, 1)
plt.scatter(range(len(groupA)), groupA, c="r")
plt.scatter(len(groupA)//2, meanA, c="black", s=120, marker="x", label="Mean")
plt.title("Group A Scatter Plot")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()

# Plot 2: Group B
plt.subplot(1, 2, 2)
plt.scatter(range(len(groupB)), groupB, c="g")
plt.scatter(len(groupB)//2, meanB, c="black", s=120, marker="x", label="Mean")
plt.title("Group B Scatter Plot")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.violinplot([groupA, groupB], showmeans=True, showmedians=True)
plt.xticks([1,2], ["Group A", "Group B"])
plt.title("Violin Plot Showing Central Tendency")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given endpoints
x0, y0 = 0, 0
x1, y1 = 6, 18

# Calculate slope
m = (y1 - y0) / (x1 - x0)

# Lists to store points
x_values = []
y_values = []

# Compute y for each x from 0 to 6
for x in range(x0, x1 + 1):
    y = m * x
    x_values.append(x)
    y_values.append(y)
    print(f"x = {x}, y = {y}")

# Plot the line
plt.plot(x_values, y_values, marker='o',linestyle="--")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line from (0,0) to (6,18)")
plt.xticks(x_values)
plt.xticks(y_values)
plt.grid(True)
plt.show()
```
