```python
import matplotlib.pyplot as plt
a = [1,2,3,4]
b = [2,4,6,8]
plt.plot(a, b)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()
```

```python
import pandas as pd
a = [1,2,3,4]
b = [2,4,6,8]

ab=pd.DataFrame({"X":a, "Y":b})

sns.lineplot(x="X",y="Y",data=ab)
plt.show()
```
